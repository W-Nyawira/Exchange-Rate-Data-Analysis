```python
#import the necessary libraries

import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
#Read the CSV file into a dataframe

df=pd.read_csv(r"C:\Users\Administrator\Downloads\Compressed\data_exam_part_II\rand-dollar.csv", index_col=0)
df.tail()
```

```python
#Splitting the data into target and feature variables
y=df['ZAR/USD']
X=df[['Value of Exports (ZAR)']]

X.head()
```

```python
#splitting the features into train and test data 
X_train, X_test,y_train, y_test= train_test_split(X,y, test_size=0.2, shuffle=False)

#fitting the model

model=LinearRegression()
model.fit(X_train, y_train)

#Getting the intercept
model_intercept=model.intercept_

print("The model's intercept is:", model_intercept)

```

```python
#Finding the coefficient of the model
slope = model.coef_[0]

print("The value of the slope of the model is:", slope)
```

```python
#Prediciting the value of the exchange rate in a month where exports total R100 000

predicted_value= slope*100000+ model_intercept
print("The predicted exchange value for R100000 exports is:", predicted_value)
```

```python
#Finding the mean squared error for the test set
from sklearn.metrics import mean_squared_error

y_pred= model.predict(X_test)

error= mean_squared_error(y_test, y_pred)

print("The mean squared error for the model is:", error)
```

```python
#Getting the R2 Value of the model
from sklearn.metrics import r2_score

r2= r2_score(y_test, y_pred)

print("The variance in the dependent variable that is predicted by the independent variable is:", r2)
```
